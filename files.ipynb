```python
from os import listdir
import os
import xarray as xr
```

```python
#path = './training/' #972 f 918i/54o
#path = './test/' #213 f 168i/45o
path = './validation/' #214 f 203i/11o
out = 0
inf = 0
flist = []

for f in listdir(path):
    ds = xr.open_mfdataset(path+f)
    if ((listdir(path).index(f)+1)<len(listdir(path))):
        nf = listdir(path).__getitem__(listdir(path).index(f)+1)
    dsn = xr.open_mfdataset(path+nf)
    #pegar o dia anterior ao de uma chuva e o dia de chuva
    #if( "rain" in ds.tags or "hail" in ds.tags or "storm" in ds.tags or "downpour" in ds.tags):
    if ((not ds.tags == "snow" and not ds.tags == "") or (not dsn.tags == "snow" and not dsn.tags == "")):
        inf = inf + 1
        flist.append(f)
        print ("ENTRA  " + f + " " + ds.tags)
    else:
        out =  out + 1
        print ("FOOOORA  " + f + " " + ds.tags)
        ds.close()
        os.remove(path+f)
        print ("removido")


print (out)
print (inf)
print (len(flist))
```
